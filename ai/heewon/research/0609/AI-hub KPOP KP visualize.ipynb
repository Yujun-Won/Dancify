```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import cv2
```

```python
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
GREEN = (0, 255, 0)
RED = (255, 0, 0)
BLUE = (0, 0, 255)
```

```python
CLASS_KP = ['c_hip', 
            'l_hip', 'l_knee', 'l_ankle', 'l_big_toe', 'l_little_toe', 
            'r_hip', 'r_knee', 'r_ankle', 'r_big_toe', 'r_little_toe', 
            'waist', 'chest', 'neck', 
            'l_shoulder', 'l_elbow', 'l_whist', 'l_thumb', 'l_little_finger',
            'r_shoulder', 'r_elbow', 'r_whist', 'r_thumb', 'r_little_finger',
           'nose', 'l_eye', 'r_eye', 'l_ear', 'r_ear']
len(CLASS_KP)
```

```python
def parse_dance_json(path_json):
    label = {}
    #print('path_json', path_json)    
    
    label['path_img'] = path_json[:-5] + '.jpg'
    
    f = open(path_json, encoding='utf-8')
    data = json.load(f)
    #print('key', data.keys())
    annotations = data['annotations']
    #print('annotations', len(annotations))
    for annotation in annotations:
        bbox = annotation['bbox']        
        keypoints = annotation['keypoints']
        keypoints3d = annotation['keypoints3d']
        #print('bbox', bbox)
        #print('keypoints', len(keypoints), keypoints)
        #print('keypoints3d', len(keypoints3d), keypoints3d)
        
        label['bbox'] = bbox
        label['kp2d'] = keypoints
        label['kp3d'] = keypoints3d
        
    return label
```

```python
def parse_dance_folder(folder_dance):
    
    paths_img = glob(folder_dance + '*.jpg')
    paths_json = glob(folder_dance + '*.json')
    print('len', len(paths_img), len(paths_json))
    
    labels = []
    for path_json in paths_json:
        label = parse_dance_json(path_json)
        labels.append(label)           
        
    return labels

labels = parse_dance_folder(folder_dance)
```

```python
def cv2_line(img, c0, c1, color, thickness):
    #print('c0, c1', c0, c1)
    if c0[2] > 1 and c1[1] > 1:        
        cv2.line(img, c0[:2], c1[:2], color, thickness)
        
def draw_heatmap_line(img, kp_centers):
    c_hip = kp_centers[0]
    l_hip = kp_centers[1]
    l_knee = kp_centers[2]
    l_ankle = kp_centers[3]
    l_big_toe = kp_centers[4]
    l_little_toe = kp_centers[5]
    r_hip = kp_centers[6]
    r_knee = kp_centers[7]
    r_ankle = kp_centers[8]
    r_big_toe = kp_centers[9]
    r_little_toe = kp_centers[10]
    
    waist = kp_centers[11]
    chest = kp_centers[12]
    neck = kp_centers[13]
    
    l_shoulder = kp_centers[14]
    l_elbow = kp_centers[15]
    l_whist = kp_centers[16]
    l_thumb = kp_centers[17]
    l_little_finger = kp_centers[18]
    
    r_shoulder = kp_centers[19]
    r_elbow = kp_centers[20]
    r_whist = kp_centers[21]
    r_thumb = kp_centers[22]
    r_little_finger = kp_centers[23]
    
    nose = kp_centers[24]
    l_eye = kp_centers[25]
    r_eye = kp_centers[26]
    l_ear = kp_centers[27]
    r_ear = kp_centers[28]
        
    cv2_line(img, c_hip, l_hip, RED, thickness=2)
    cv2_line(img, l_hip, l_knee, RED, thickness=2)
    cv2_line(img, l_knee, l_ankle, RED, thickness=2)
    cv2_line(img, l_ankle, l_big_toe, RED, thickness=2)
    cv2_line(img, l_ankle, l_little_toe, RED, thickness=2)
    
    cv2_line(img, c_hip, r_hip, BLUE, thickness=2)
    cv2_line(img, r_hip, r_knee, BLUE, thickness=2)
    cv2_line(img, r_knee, r_ankle, BLUE, thickness=2)
    cv2_line(img, r_ankle, r_big_toe, BLUE, thickness=2)
    cv2_line(img, r_ankle, r_little_toe, BLUE, thickness=2)
    
    cv2_line(img, c_hip, waist, WHITE, thickness=2)    
    cv2_line(img, waist, chest, WHITE, thickness=2)
    cv2_line(img, chest, neck, WHITE, thickness=2)
    
    cv2_line(img, neck, l_shoulder, RED, thickness=2)
    cv2_line(img, l_shoulder, l_elbow, RED, thickness=2)
    cv2_line(img, l_elbow, l_whist, RED, thickness=2)
    cv2_line(img, l_whist, l_thumb, RED, thickness=2)
    cv2_line(img, l_whist, l_little_finger, RED, thickness=2)
    
    cv2_line(img, neck, r_shoulder, BLUE, thickness=2)
    cv2_line(img, r_shoulder, r_elbow, BLUE, thickness=2)
    cv2_line(img, r_elbow, r_whist, BLUE, thickness=2)
    cv2_line(img, r_whist, r_thumb, BLUE, thickness=2)
    cv2_line(img, r_whist, r_little_finger, BLUE, thickness=2)
    
    cv2_line(img, neck, nose, WHITE, thickness=1)    
    cv2_line(img, nose, l_eye, RED, thickness=1)
    cv2_line(img, l_eye, l_ear, RED, thickness=1)
    cv2_line(img, nose, r_eye, BLUE, thickness=1)
    cv2_line(img, r_eye, r_ear, BLUE, thickness=1)
    
    return img
```

```python
def display_kp_on_image(label):
    path_img = label['path_img']
    bbox = label['bbox']
    kp = label['kp2d']
    kp3d = label['kp3d']
    
    kp_2d = np.array(kp).reshape((-1, 3))
    kp_3d = 300 - np.abs(np.array(kp3d)).reshape((-1, 3)).astype(int)
    kp_3d *= 2
    
    if not os.path.exists(path_img):
        print('not exist', path_img)
        return
    
    arr = cv2.imread(path_img)
    
    x0, y0, w, h = bbox
    x1 = x0 + w
    y1 = y0 + h
    
    cv2.putText(arr, path_img, (0, 20), cv2.FONT_HERSHEY_PLAIN, 1, WHITE, 1)
    cv2.rectangle(arr, (x0, y0), (x1, y1), GREEN, 2)
    
    radius = 3
    for i in range(len(kp_2d)):
        kp_cls = CLASS_KP[i]
        x, y, visible = kp_2d[i]
        
        thickness = min(2, visible)
        color = WHITE        
        if kp_cls.startswith('l_'):
            color = RED
        elif kp_cls.startswith('r_'):
            color = BLUE
        cv2.circle(arr, (x, y), radius, color, thickness)       
    
    draw_heatmap_line(arr, kp_2d)
    draw_heatmap_line(arr, kp_3d)
    return arr
```

```python
out = cv2.VideoWriter('Kpop_KP3D_visualize.avi',cv2.VideoWriter_fourcc(*'DIVX'), 30, (1920, 1080))
```

```python
for label in labels:
    visualized_arr = display_kp_on_image(label)
    cv2.imshow('', visualized_arr)
    ret = cv2.waitKey(1)
    
    #print('pressed key is {0}'.format(ret))
    out.write(visualized_arr)
    
out.release()
```
